```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
im=plt.imread('../../adv-attacks-pytorch-101/imagenet/input.png')
plt.imshow(im)
```

```python
class DominantColor():
    def __init__(self,path,n):
        self.path=path
        self.get_image()
        self.yuv=np.array([[0.299,0.587,0.114],[-0.14713,-0.28888,0.436],[0.615,-0.5149,-0.10001]])
        self.rgb=m_inv=np.linalg.inv(self.yuv)
        self.km=0
        self.colors=0
        self.n=n
    def get_image(self):
        self.im=plt.imread(self.path)
    def show_image(self):
        plt.figure(figsize=(5,8))
        plt.axis('off')
        plt.imshow(self.im)
    
    
    def fit(self,):
        im=self.im.reshape(-1,3)
        im=self.yuv.dot(im.T).T   # (h*w,3) in yuv plane
        self.km=KMeans(n_clusters=self.n)
        self.km.fit(im)
        self.colors=self.rgb.dot(self.km.cluster_centers_.T).T
        
    def show_colors(self,save=False):
        n_centers,n_labels=np.unique(self.km.labels_,return_counts=True)
        n_labels=n_labels/ float(sum(n_labels))
        temp=np.zeros((1200,2000,3))
        wid=0
        for ix in range(self.n):
            end=int(temp.shape[1]*n_labels[ix])
            temp[:,wid:wid+end,:]=self.colors[ix]
            wid+=end
        #plt.figure(figsize=(5,15))
        plt.axis('off')
        plt.imshow(temp)
        if save:
            plt.savefig("colors.jpeg")
```

```python
m=DominantColor('../../adv-attacks-pytorch-101/imagenet/input.png',3)
```

```python
m.fit()
```

```python
m.show_colors(save=True)
```
